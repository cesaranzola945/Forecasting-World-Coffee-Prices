# Import Brazil Weather based on 

```python
# import brazil data downloaded using google earth
folder_path = "C:/Users/canzolbr/Documents/GitHub/Forecasting-World-Coffee-Prices/Data/Weather/google-earth/Brazil Extended/"
```

```python
import pandas as pd
import os
os.chdir(folder_path)
```

```python
dir_list = os.listdir()
df = []
for file in dir_list:
    d = pd.read_csv(folder_path + file)
    df.append(d)
```

```python
df_all = pd.concat(df,axis=0)
```

```python
df_all = df_all.reset_index(drop=True)
df_all = df_all.drop('system:index',axis=1)
```

```python
# brazilian states
df_all['state_name'].unique()
```

```python
# create date vector
df_all['year'] = pd.to_numeric(df_all['year'])
df_all['month'] = pd.to_numeric(df_all['month'])
```

```python
df_all['date'] = pd.to_datetime(dict(year = df_all['year'],
                                     month = df_all['month'], 
                                     day = 1))
```

```python
df_all = df_all.set_index('date')
```

#### Check data for minas gerais - largest coffee producer estate in brazil

```python
df_mg = df_all[df_all['state_name']=='Minas Gerais']
```
