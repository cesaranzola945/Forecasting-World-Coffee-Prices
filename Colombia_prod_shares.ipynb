# Colombia department production shares

```python
eva_path = "//Wcalisrv/cearn/Datos/Agropecuario/EVA/datos/modificado/"

output_path = "C:/Users/canzolbr/Documents/GitHub/Forecasting-World-Coffee-Prices/Data/"
```

```python
import pandas as pd
df = pd.read_stata(eva_path + "eva_mpio_19-24.dta",convert_categoricals=False)
```

```python
df['cultivo'].unique()
```

```python
df_c = df[df['cultivo']=='Café']
```

```python
df_c_sum = df_c[['coddpto','depto','produccion']].groupby(['coddpto','depto']).sum()
```

```python
df_c_sum['prod_share'] =  df_c_sum['produccion']/df_c_sum['produccion'].sum()
df_c_sum = df_c_sum.reset_index()
```

```python
df_c_sum = df_c_sum[['coddpto','depto','prod_share']]
```

```python
df_c_sum.to_csv(output_path + "/col_prod_shares.csv")
```
