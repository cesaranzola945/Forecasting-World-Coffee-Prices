# Brazil Production Shares

Data source from the following Brazilian source:
[https://www.ibge.gov.br/explica/producao-agropecuaria/cafe/br]

```python
local_link = "C:/Users/canzolbr/Documents/GitHub/Forecasting-World-Coffee-Prices/Data/"
```

```python
import pandas as pd 
df = pd.read_csv(local_link + "brazil_coffee_productvalue.csv",encoding='utf-8',skiprows=1)
```

```python
df['valor']
```

```python
# simply convert string to numeric value
df['valor']  = df['valor'].str.replace('-','')
df['valor'] = df['valor'].str.replace('.', '')
df['valor'] = pd.to_numeric(df['valor']) 
```

```python
df = df[['localidade','valor']]
df = df.rename(columns={'localidade':'division','valor':'prod_val'})
```

```python
df['prod_share'] = df['prod_val']/df['prod_val'].sum()
```

```python
!pip install unidecode
```

```python
df = df.dropna(subset=['division'])
```

```python
from unidecode import unidecode
df['state_name'] = df['division'].apply(unidecode)
```

```python
df = df.dropna(subset=['prod_share'])
```

```python
df.sort_values('prod_share',ascending=False)
```
